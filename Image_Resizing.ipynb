Image resizing is not only for enlarging an image but it can also be used to make an image smaller

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from google.colab.patches import cv2_imshow



img= cv2.imread('image1.jpg')
cv2_imshow(img)
print(img.shape)


#the value of fx and fy determines the new size of the image, for this case, it increases the size of the image by 10 times 


#Nearest method to resize

near_img = cv2.resize(img,None,fx = 10, fy=10, interpolation = cv2.INTER_NEAREST) 
print ('Nearest Image Size:',near_img.shape)
cv2_imshow(near_img)
#cv2.imwrite('NewNameforNearImage.jpg',near_img)   for saving with cv2, we use this


#Bilinear Method to resize

bilinear_img= cv2.resize(img,None,fx = 10, fy=10, interpolation = cv2.INTER_LINEAR)
cv2_imshow(bilinear_img)
#cv2.imwrite('NewNameforBilinearImage.jpg',bilinear_img)
print('Bilinear Image Size:',bilinear_img.shape)


#Bicubic Method to Resize

bicubic_img= cv2.resize(img,None,fx=10,fy=10,interpolation = cv2.INTER_CUBIC)
cv2_imshow(bicubic_img)
#cv2.imwrite('NewNameforCubicImage.jpg',bicubic_img)
print('Bicubic Image Size:',bicubic_img.shape)


#Displaying all three resized images in a single row:

fig, axess = plt.subplots ( ncols=3, figsize= (7,8)) #rows is the number of images 

ax3,ax4,ax5=axess
plt.gray()

ax3.imshow(near_img[:,:,::-1])
ax4.imshow(bilinear_img[:,:,::-1])
ax5.imshow(bicubic_img[:,:,::-1])

plt.show()
```
